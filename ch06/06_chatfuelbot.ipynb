# Chatfuelを用いたピザ注文ボット

Chatfuelの公式サイトにアクセスすると、「Try it free」ボタンがあります。クリックすると、Facebookアカウントでのログインを求められるので、必要な権限を与えてください。

## ボットの作成

ログインすると、簡単なチュートリアルが表示されます。ここでは、与えられた選択肢の中からユーザーが選択するだけのシンプルなチャットボットを作成します。chatfuelを使用して作成されたボットにはAIやNLPが搭載されていますが、ここでの説明の範囲を超えているため、これらについてはドキュメントで確認してください。

これはチャットボットの最もシンプルな形態で、ユーザーはボットと実際に会話するのではなく（テキスト）、ボットが提供する選択肢の中から1つを選ぶだけであることに注意してください。

ここでは、大量の「ブロック」を使用します。基本的にブロックは、ボットのスクリプトの一部であり、ユーザーが選択したオプションに基づいてボットがナビゲートします。

image.png

この基本的な例では、chatfuelを使って、ピザを注文するボットを作っています。ピザのサイズはS、M、Lの3種類、トッピングはオニオン、エキストラチーズ、両方の3種類から選択できます。

上の画像では、ユーザーがチャットボットと会話を始めたときに、チャットボットが最初に送るウェルカムメッセージを見ることができます。テキストボックス内のテキストは、お客様のご要望に応じて変更可能です。

上の画像では、青色の部分がユーザーに表示されるオプションを表し、グレーの部分がユーザーがそのオプションを選択したときにボットがジャンプするブロックを表しています。

例えば、ユーザーが「ピザを注文する（Order Pizza）」を選択すると、ボットは「サイズ」ブロックにジャンプし、以下のように表示されます。

image.png

同様に、ボットの反応を変えるオプションの数が多ければ、ブロックの数も増えます。オプションを使った適切なリダイレクションや、redirection要素を使った明示的なリダイレクションにより、ブロック間を行き来することができます。その他、「タイピング」、「画像」、「クイックリプライ」などの要素を使用して、ユーザー体験を向上させることができます。

今回のケースでは、3つのサイズのピザがあり、それぞれのサイズに3つのトッピングの選択肢があるので、すべての異なる結果を処理するために少なくとも9つのブロックが必要です。また、その他のケースを処理するためにいくつかの追加ブロックが必要です。
